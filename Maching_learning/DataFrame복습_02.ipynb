```python
# 그룹변수가 하나가 아닌, 전체 연속형 변수에 대한 집계
abalone_df.groupby(abalone_df['sex']).mean()
```

```python
# 다음과 같이 간단하게 표현 가능.!
abalone_df.groupby('sex').mean()
```

```python
# 새로운 조건에 맞는 col 변수 추가.
abalone_df['length_bool'] = np.where(abalone_df['length'] > abalone_df['length'].median(),
                                     'length_long', 
                                     'length_short')
abalone_df
```

```python
# 그룹변수를 2개이상 선택해서 총계처리는 어떻게 할까.? => EDA (EDA(Exploratory Data Analysis, 탐색적 데이터 분석))
abalone_df.groupby(['sex', 'length_bool']).mean()
```

```python
# 그룹변수 2개이상 선택해서 총계처리후 columns 열만 출력.
abalone_df.groupby(['sex','length_bool'])['whole_weight'].mean()
```

```python
# *** 사용자 함수를 통한 집계 *******
import math

def avg_ceil(x, y, z):
  return math.ceil((x+y+z)) # ceil 올림.

abalone_df[['diameter', 'height', 'whole_weight']].apply(lambda x : avg_ceil(x[0], x[1], x[2]), axis = 1)
```

```python
# 문제
# 1. 사용자 정의 함수 정의
# 2. ['diameter', 'height', 'whole_weight'] 변수 사용
# 3. 세 변수의 합이 1을 넘으면 True, 아니면 False 출력 후 answer 변수에 저장
# 4. abalone_df에 answer 열을 추가하고 입력
```

```python
def f(x, y, z):
  answer = x + y + z
  if answer > 1:
    answer = True
    return answer
  else:
    answer = False
    return answer
  
abalone_df['answer'] = abalone_df[['diameter', 'height', 'whole_weight']].apply(lambda x : f(x[0], x[1], x[2]), axis = 1)
abalone_df[['diameter', 'height', 'whole_weight', 'answer']]
```
