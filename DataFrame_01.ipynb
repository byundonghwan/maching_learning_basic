```python
# Series 실습
pd.Series([1,2,3]) # 인덱스가 자동으로 제공한다, 연산시 numpy보다 연산속도가 느리다. 

# 데이터프레임(집합) => 데이터프레임(집합) <=> 시리즈(단일)
# 데이터프레임과 시리즈는 동일하구나.
series = df['name']
series
```

```python
# pd.Series 함수를 통해서 데이터프레임 만드는 것도 가능
series = pd.Series([1,2,3,4])
series
```

```python
# pd.Series의 옵션이 무엇이 있는지 확인
# index => 할당값의 지정 이름, 중복 가능
# dtype => 대표적 : int, float, string, boolean 등 사용.
series = pd.Series([10,2,35,4], index = ['a', 'b', 'c', 'd'], dtype = float)
series
```

```python
# 시리즈 정렬 
series = series.sort_values(ascending = True)
series
```
